```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def stem(text):
    """
    Stems a list of words using PorterStemmer.

    Args:
        text (list): A list of words to be stemmed.

    Returns:
        str: A string containing the stemmed words joined by spaces.
    """
    if isinstance(text, list):  # Check if text is a list
        y = []
        for i in text:  # Iterate over words in the list
            if isinstance(i, str):  # Check if the element is a string
                y.append(ps.stem(i))
            else:
                # Handle non-string elements in the list, e.g., skip them
                pass
        return " ".join(y)
    else:
        # Handle cases where text is not a list, e.g., a single string
        return " ".join([ps.stem(i) for i in text.split()])

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]


    for i in movies_list:
        print(new_df.iloc[i[0]].title)

```
